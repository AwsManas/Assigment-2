```python
from math import e

def pandas_entropy(column, base=None):
    vc = pd.Series(column).value_counts(normalize=True, sort=False)
    base = e if base is None else base
    return -(vc * np.log(vc)/np.log(base)).sum()
```

```python
def whichCategory(category):
    # accepts a list of categories and returns the final category
    x = category[0]
    restr = ['Restaurant','Food','Bar','Steak','Sandwich','Chicken','Kitchen','Barbecue','Bakery','Brew','Pub','Diner']
    medical = ['Hospital','Doctor','Nurse','Medical','Clinic', 'Therap','Health','Physical','Psycho','Mental','Emergency','Surg','Gynec']
    legalAdmin = ['Legal','Attorney','Law']
    school = ['School','Book','Toy','Kinder']
    entertainment = ['Sport','Bowling','Arcade','Play','Movie','Tele','Sport','Golf','Party','Comedy','Field']
    citizenLocs = ['Office','Bank','Agency','Social']
    venues = ['Hall','Banquet','Catering','Hotel','Resort','Mall','Spa','Gallery','Market']
    ShopsStores = ['Shop','Store','Dealer']
    assocOrgs = ['Association','Organization','Union','Service']
    if any(map(x.__contains__, restr)):
        return 'Restaurant'
    elif any(map(x.__contains__, legalAdmin)):
        return 'Legal'
    elif any(map(x.__contains__, school)):
        return 'School'
    elif any(map(x.__contains__, entertainment)):
        return 'Entertainment'
    elif any(map(x.__contains__, medical)):
        return 'Medical'
    elif any(map(x.__contains__, citizenLocs)):
        return 'Public'
    elif any(map(x.__contains__, venues)):
        return 'Venues'
    elif any(map(x.__contains__, ShopsStores)):
        return 'ShopsStores'
    elif any(map(x.__contains__, assocOrgs)):
        return 'AssocOrgs'
    else:
        return 'Others'
```

```python
def encode_and_bind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]])
    res = pd.concat([original_dataframe, dummies], axis=1)
    res = res.drop([feature_to_encode], axis=1)
    return(res) 
```

```python
def MSE(predictions, labels):
    differences = [(x-y)**2 for x,y in zip(np.array(predictions),np.array(labels))]
    return sum(differences) / len(differences)
```

```python
data_dir = "./final_data/"
Xtrain = pd.read_pickle(data_dir+"X_train.pkl")
Xval = pd.read_pickle(data_dir+"X_val.pkl")
Xtest = pd.read_pickle(data_dir+"X_test.pkl")
Ytrain = pd.read_pickle(data_dir+"Y_train.pkl")
Yval = pd.read_pickle(data_dir+"Y_val.pkl")
Ytest = pd.read_pickle(data_dir+"Y_test.pkl")
```

```python
Xtrain.drop(['gPlusUserId','gPlusPlaceId'],axis=1,inplace = True)
Xval.drop(['gPlusUserId','gPlusPlaceId'],axis=1,inplace = True)
Xtest.drop(['gPlusUserId','gPlusPlaceId'],axis=1,inplace = True)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 100, min_samples_split = 250,random_state = 42)
rf.fit(Xtrain, Ytrain)
```

```python
rf_pred_val = rf.predict(Xval)
```

```python
validMSE = MSE(rf_pred_val,Yval)
```

```python
validMSE
```

```python
rf_pred_test = rf.predict(Xtest)
testMSE = MSE(rf_pred_test,Ytest)
```
