```python
usa = usa.drop(usa[usa.unixReviewTime.isna()].index)
usa.head(2)
```

```python
usa['date'] = pd.to_datetime(usa['unixReviewTime'],unit='s')
usa['month'] = pd.to_datetime(usa['unixReviewTime'],unit='s').dt.month
usa['year'] = pd.to_datetime(usa['unixReviewTime'],unit='s').dt.year
usa['month-year']  = pd.to_datetime(usa['unixReviewTime'],unit='s').dt.strftime('%Y-%m')
```

```python
t = pd.DataFrame()
fig = plt.figure(figsize=(16, 8))
for name in tqdm(common_users):
    df_usr = usa[ usa.userName == name]
    imp = pd.DataFrame()
    imp['rating'] = df_usr['rating']
    imp['time'] = df_usr['month-year']
    imp.reset_index(drop=True)
    to_plot = pd.DataFrame()
    to_plot = imp.groupby('time').mean()
    to_plot = to_plot.reset_index()
    to_plot = to_plot.sort_values(by=['time'])
    to_plot['time']= pd.to_datetime(to_plot['time'])
    plt.plot(list(to_plot['time']), list(to_plot['rating']), label=name)
plt.xticks(rotation=90)
plt.xlim([ datetime.datetime(2012, 5, 1) , datetime.datetime(2014, 4, 30) ])
plt.legend()
plt.show()
```

```python
y = [0]*12
x = list(range(1,13))
s = [0]*12
t = pd.DataFrame()
fig = plt.figure(figsize=(10, 8))
for name in tqdm(common_users):
    df_usr = usa[ usa.userName == name]
    imp = pd.DataFrame()
    imp['rating'] = df_usr['rating']
    imp['time'] = df_usr['month']
    imp.reset_index(drop=True)
    to_plot = pd.DataFrame()
    to_plot = imp.groupby('time').mean()
    to_plot = to_plot.reset_index()
    to_plot = to_plot.sort_values(by=['time'])
    list_mnth = list(to_plot['time'])
    list_rating = list(to_plot['rating'])
    
    for(m,r) in zip(list_mnth,list_rating):
        #print(m,r)
        s[m-1]+=1
        y[m-1]+=r

for i in range(len(y)):
    y[i] = y[i]/s[i]
    

ig = plt.figure(figsize = (10, 5))
plt.bar(x, y, color ='maroon',width = 0.4)
plt.show()
```

```python
user_freq_ = dict(usa['userName'].value_counts())
common_users = []
tlow = 500
for u in user_freq_:
    if user_freq_[u] < tlow:
        break
    common_users.append(u)    
```
