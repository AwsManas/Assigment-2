```python
from metrics import MAE,MSE
from sklearn.linear_model import LinearRegression
```

```python
y_ = new_df['rating']
y_.head()
```

```python
X_ = new_df.drop(labels = ['rating'],axis = 1)
```

```python
X_.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_, y_, test_size=0.2, shuffle=False) 
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, shuffle = False)
```

```python
X_train.to_pickle("../Datasets/temporal/final/X_train.pkl")
X_val.to_pickle("../Datasets/temporal/final/X_val.pkl")
X_test.to_pickle("../Datasets/temporal/final/X_test.pkl")
```

```python
dataset_dir = "../Datasets/temporal/final/"
np_y = pd.DataFrame(np.array([y_train]).T,columns = ["rating"])
np_y.to_pickle(dataset_dir+"y_train.pkl")

np_y = pd.DataFrame(np.array([y_test]).T,columns = ["rating"])
np_y.to_pickle(dataset_dir+"y_test.pkl")


np_y = pd.DataFrame(np.array([y_val]).T,columns = ["rating"])
np_y.to_pickle(dataset_dir+"y_val.pkl")
```

```python
X_train = X_train.drop(labels = ['gPlusUserId', 'gPlusPlaceId'],axis = 1).to_numpy()
X_val = X_val.drop(labels = ['gPlusUserId', 'gPlusPlaceId'],axis = 1).to_numpy()
X_test = X_test.drop(labels = ['gPlusUserId', 'gPlusPlaceId'],axis = 1).to_numpy()
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()
y_val = y_val.to_numpy()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_val = lr.predict(X_val)
print("Validation")
print(MAE(predictions=y_pred_val, labels=y_val))
print(MSE(predictions=y_pred_val, labels=y_val))
print("Testing")
y_pred_ts = lr.predict(X_test)
print(MAE(predictions=y_pred_ts, labels=y_test))
print(MSE(predictions=y_pred_ts, labels=y_test))
```

```python
best_model = None
best_mae = np.Inf
best_mse = np.Inf
best_alpha = np.Inf
print("Validation")
for alpha in [1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3]:
    r = Ridge(alpha=alpha)
    r.fit(X_train, y_train)
    y_pred_val = r.predict(X_val)
    mae_val = MAE(predictions=y_pred_val, labels=y_val)
    mse_val = MSE(predictions=y_pred_val, labels=y_val)
    print(f"Alpha: {alpha}, MAE: {mae_val} , MSE : {mse_val}")
    
    if mae_val < best_mae:
        best_mae = mae_val
        best_model = r
        best_alpha = alpha
```

```python
best_alpha
```

```python
y_pred = best_model.predict(X_test)
```

```python
print(MAE(predictions=y_pred, labels=y_test))
print(MSE(predictions=y_pred, labels=y_test))
```
